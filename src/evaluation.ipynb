```python
def cumulative_violations(scenario, approach, params):
    working_dir = env_vars.PROJ_HOME + 'results/' + str(params) + '_params/' + scenario + '/' + approach + '/'
    iterations_dir = [d for d in os.listdir(working_dir) if os.path.isdir(working_dir + d)]
    iterations = {}
    for iter_folder in iterations_dir:
        try:
            iterations[iter_folder] = {'total_violations': 0}
            violation_scores = pd.read_csv(working_dir  + iter_folder + '/' + 'violation_scores_' + iter_folder + '.csv', index_col=0)
            violations = violation_scores[scenario].to_list()
            for v in violations:
                iterations[iter_folder]['total_violations'] += max(0, v)
        except Exception as e:
            print(e)

    # total_violation_score_cac_adv = 0
    scores = []
    for iteration in iterations:
        # total_violation_score_cac_adv += iterations[cac_iter]['total_violations']
        scores.append(iterations[iteration]['total_violations'])
    # total_violation_score_cac_adv

    cum_scores = []
    for i in range(1, len(scores)+1): cum_scores.append(sum(scores[:i]))
    return cum_scores
```

```python
def calculate_final_cumulative_score(cum_scores):
    """
    Calculate the final cumulative score from the cumulative scores list.

    Parameters:
    - cum_scores: List of cumulative scores (list of integers).

    Returns:
    - final_cum_score: Final cumulative score (integer).
    """
    return cum_scores[-1] if cum_scores else 0

# RTL1 Evaluation
rtl1_rand_5p = cumulative_violations('A.RTL1', 'rand', 5)
rtl1_rand_5p_final = calculate_final_cumulative_score(rtl1_rand_5p)

rtl1_ga_5p = cumulative_violations('A.RTL1', 'ga', 5)
rtl1_ga_5p_final = calculate_final_cumulative_score(rtl1_ga_5p)

rtl1_drl_5p = cumulative_violations('A.RTL1', 'drl', 5)
rtl1_drl_5p_final = calculate_final_cumulative_score(rtl1_drl_5p)

rtl1_rand_10p = cumulative_violations('A.RTL1', 'rand', 10)
rtl1_rand_10p_final = calculate_final_cumulative_score(rtl1_rand_10p)

rtl1_ga_10p = cumulative_violations('A.RTL1', 'ga', 10)
rtl1_ga_10p_final = calculate_final_cumulative_score(rtl1_ga_10p)

rtl1_drl_10p = cumulative_violations('A.RTL1', 'drl', 10)
rtl1_drl_10p_final = calculate_final_cumulative_score(rtl1_drl_10p)

rtl1_rand_15p = cumulative_violations('A.RTL1', 'rand', 15)
rtl1_rand_15p_final = calculate_final_cumulative_score(rtl1_rand_15p)

rtl1_ga_15p = cumulative_violations('A.RTL1', 'ga', 15)
rtl1_ga_15p_final = calculate_final_cumulative_score(rtl1_ga_15p)

rtl1_drl_15p = cumulative_violations('A.RTL1', 'drl', 15)
rtl1_drl_15p_final = calculate_final_cumulative_score(rtl1_drl_15p)

rtl1_rand_20p = cumulative_violations('A.RTL1', 'rand', 20)
rtl1_rand_20p_final = calculate_final_cumulative_score(rtl1_rand_20p)

rtl1_ga_20p = cumulative_violations('A.RTL1', 'ga', 20)
rtl1_ga_20p_final = calculate_final_cumulative_score(rtl1_ga_20p)

rtl1_drl_20p = cumulative_violations('A.RTL1', 'drl', 20)
rtl1_drl_20p_final = calculate_final_cumulative_score(rtl1_drl_20p)

# RTL2 Evaluation
rtl2_rand_5p = cumulative_violations('A.RTL2', 'rand', 5)
rtl2_rand_5p_final = calculate_final_cumulative_score(rtl2_rand_5p)

rtl2_ga_5p = cumulative_violations('A.RTL2', 'ga', 5)
rtl2_ga_5p_final = calculate_final_cumulative_score(rtl2_ga_5p)

rtl2_drl_5p = cumulative_violations('A.RTL2', 'drl', 5)
rtl2_drl_5p_final = calculate_final_cumulative_score(rtl2_drl_5p)

rtl2_rand_10p = cumulative_violations('A.RTL2', 'rand', 10)
rtl2_rand_10p_final = calculate_final_cumulative_score(rtl2_rand_10p)

rtl2_ga_10p = cumulative_violations('A.RTL2', 'ga', 10)
rtl2_ga_10p_final = calculate_final_cumulative_score(rtl2_ga_10p)

rtl2_drl_10p = cumulative_violations('A.RTL2', 'drl', 10)
rtl2_drl_10p_final = calculate_final_cumulative_score(rtl2_drl_10p)

rtl2_rand_15p = cumulative_violations('A.RTL2', 'rand', 15)
rtl2_rand_15p_final = calculate_final_cumulative_score(rtl2_rand_15p)

rtl2_ga_15p = cumulative_violations('A.RTL2', 'ga', 15)
rtl2_ga_15p_final = calculate_final_cumulative_score(rtl2_ga_15p)

rtl2_drl_15p = cumulative_violations('A.RTL2', 'drl', 15)
rtl2_drl_15p_final = calculate_final_cumulative_score(rtl2_drl_15p)

rtl2_rand_20p = cumulative_violations('A.RTL2', 'rand', 20)
rtl2_rand_20p_final = calculate_final_cumulative_score(rtl2_rand_20p)

rtl2_ga_20p = cumulative_violations('A.RTL2', 'ga', 20)
rtl2_ga_20p_final = calculate_final_cumulative_score(rtl2_ga_20p)

rtl2_drl_20p = cumulative_violations('A.RTL2', 'drl', 20)
rtl2_drl_20p_final = calculate_final_cumulative_score(rtl2_drl_20p)

# RTL3 Evaluation (only 20 params)
rtl3_rand_20p = cumulative_violations('A.RTL3', 'rand', 20)
rtl3_rand_20p_final = calculate_final_cumulative_score(rtl3_rand_20p)

rtl3_ga_20p = cumulative_violations('A.RTL3', 'ga', 20)
rtl3_ga_20p_final = calculate_final_cumulative_score(rtl3_ga_20p)

rtl3_drl_20p = cumulative_violations('A.RTL3', 'drl', 20)
rtl3_drl_20p_final = calculate_final_cumulative_score(rtl3_drl_20p)

# AUTO1 Evaluation (only 20 params)
auto1_rand_20p = cumulative_violations('A.AUTO1', 'rand', 20)
auto1_rand_20p_final = calculate_final_cumulative_score(auto1_rand_20p)

auto1_ga_20p = cumulative_violations('A.AUTO1', 'ga', 20)
auto1_ga_20p_final = calculate_final_cumulative_score(auto1_ga_20p)

auto1_drl_20p = cumulative_violations('A.AUTO1', 'drl', 20)
auto1_drl_20p_final = calculate_final_cumulative_score(auto1_drl_20p)

# BRAKE1 Evaluation (only 20 params)
brake1_rand_20p = cumulative_violations('A.BRAKE1', 'rand', 20)
brake1_rand_20p_final = calculate_final_cumulative_score(brake1_rand_20p)

brake1_ga_20p = cumulative_violations('A.BRAKE1', 'ga', 20)
brake1_ga_20p_final = calculate_final_cumulative_score(brake1_ga_20p)

brake1_drl_20p = cumulative_violations('A.BRAKE1', 'drl', 20)
brake1_drl_20p_final = calculate_final_cumulative_score(brake1_drl_20p)

# LAND1 Evaluation (only 20 params)
land1_rand_20p = cumulative_violations('A.LAND1', 'rand', 20)
land1_rand_20p_final = calculate_final_cumulative_score(land1_rand_20p)

land1_ga_20p = cumulative_violations('A.LAND1', 'ga', 20)
land1_ga_20p_final = calculate_final_cumulative_score(land1_ga_20p)

land1_drl_20p = cumulative_violations('A.LAND1', 'drl', 20)
land1_drl_20p_final = calculate_final_cumulative_score(land1_drl_20p)

# LAND2 Evaluation (only 20 params)
land2_rand_20p = cumulative_violations('A.LAND2', 'rand', 20)
land2_rand_20p_final = calculate_final_cumulative_score(land2_rand_20p)

land2_ga_20p = cumulative_violations('A.LAND2', 'ga', 20)
land2_ga_20p_final = calculate_final_cumulative_score(land2_ga_20p)

land2_drl_20p = cumulative_violations('A.LAND2', 'drl', 20)
land2_drl_20p_final = calculate_final_cumulative_score(land2_drl_20p)

# LOITER1 Evaluation (only 20 params)
loiter1_rand_20p = cumulative_violations('A.LOITER1', 'rand', 20)
loiter1_rand_20p_final = calculate_final_cumulative_score(loiter1_rand_20p)

loiter1_ga_20p = cumulative_violations('A.LOITER1', 'ga', 20)
loiter1_ga_20p_final = calculate_final_cumulative_score(loiter1_ga_20p)

loiter1_drl_20p = cumulative_violations('A.LOITER1', 'drl', 20)
loiter1_drl_20p_final = calculate_final_cumulative_score(loiter1_drl_20p)

# ALT_HOLD1 Evaluation (only 20 params)
# Uncomment if data is available
alt_hold1_rand_20p = cumulative_violations('A.ALT_HOLD1', 'rand', 20)
alt_hold1_rand_20p_final = calculate_final_cumulative_score(alt_hold1_rand_20p)

alt_hold1_ga_20p = cumulative_violations('A.ALT_HOLD1', 'ga', 20)
alt_hold1_ga_20p_final = calculate_final_cumulative_score(alt_hold1_ga_20p)

alt_hold1_drl_20p = cumulative_violations('A.ALT_HOLD1', 'drl', 20)
alt_hold1_drl_20p_final = calculate_final_cumulative_score(alt_hold1_drl_20p)

# ALT_HOLD2 Evaluation (only 20 params)
alt_hold2_rand_20p = cumulative_violations('A.ALT_HOLD2', 'rand', 20)
alt_hold2_rand_20p_final = calculate_final_cumulative_score(alt_hold2_rand_20p)

alt_hold2_ga_20p = cumulative_violations('A.ALT_HOLD2', 'ga', 20)
alt_hold2_ga_20p_final = calculate_final_cumulative_score(alt_hold2_ga_20p)

alt_hold2_drl_20p = cumulative_violations('A.ALT_HOLD2', 'drl', 20)
alt_hold2_drl_20p_final = calculate_final_cumulative_score(alt_hold2_drl_20p)

# Print final cumulative scores
print("RTL1 (5 params) - RAND:", rtl1_rand_5p_final, "GA:", rtl1_ga_5p_final, "DRL:", rtl1_drl_5p_final)
print("RTL1 (10 params) - RAND:", rtl1_rand_10p_final, "GA:", rtl1_ga_10p_final, "DRL:", rtl1_drl_10p_final)
print("RTL1 (15 params) - RAND:", rtl1_rand_15p_final, "GA:", rtl1_ga_15p_final, "DRL:", rtl1_drl_15p_final)
print("RTL1 (20 params) - RAND:", rtl1_rand_20p_final, "GA:", rtl1_ga_20p_final, "DRL:", rtl1_drl_20p_final)

print("RTL2 (5 params) - RAND:", rtl2_rand_5p_final, "GA:", rtl2_ga_5p_final, "DRL:", rtl2_drl_5p_final)
print("RTL2 (10 params) - RAND:", rtl2_rand_10p_final, "GA:", rtl2_ga_10p_final, "DRL:", rtl2_drl_10p_final)
print("RTL2 (15 params) - RAND:", rtl2_rand_15p_final, "GA:", rtl2_ga_15p_final, "DRL:", rtl2_drl_15p_final)
print("RTL2 (20 params) - RAND:", rtl2_rand_20p_final, "GA:", rtl2_ga_20p_final, "DRL:", rtl2_drl_20p_final)

print("RTL3 (20 params) - RAND:", rtl3_rand_20p_final, "GA:", rtl3_ga_20p_final, "DRL:", rtl3_drl_20p_final)

print("AUTO1 (20 params) - RAND:", auto1_rand_20p_final, "GA:", auto1_ga_20p_final, "DRL:", auto1_drl_20p_final)

print("BRAKE1 (20 params) - RAND:", brake1_rand_20p_final, "GA:", brake1_ga_20p_final, "DRL:", brake1_drl_20p_final)

print("LAND1 (20 params) - RAND:", land1_rand_20p_final, "GA:", land1_ga_20p_final, "DRL:", land1_drl_20p_final)

print("LAND2 (20 params) - RAND:", land2_rand_20p_final, "GA:", land2_ga_20p_final, "DRL:", land2_drl_20p_final)

print("LOITER1 (20 params) - RAND:", loiter1_rand_20p_final, "GA:", loiter1_ga_20p_final, "DRL:", loiter1_drl_20p_final)

print("ALT_HOLD1 (20 params) - RAND:", alt_hold1_rand_20p_final, "GA:", alt_hold1_ga_20p_final, "DRL:", alt_hold1_drl_20p_final)

print("ALT_HOLD2 (20 params) - RAND:", alt_hold2_rand_20p_final, "GA:", alt_hold2_ga_20p_final, "DRL:", alt_hold2_drl_20p_final)
```

```python
def plot(scenario,
        drl_5p, ga_5p, rand_5p,
        drl_10p, ga_10p, rand_10p,
        drl_15p, ga_15p, rand_15p,
        drl_20p, ga_20p, rand_20p):

    # gs = gridspec.GridSpec(1, 2) 
    fig, ax = plt.subplots(nrows=2, ncols=2)
    fig.tight_layout(pad=3)

    plt.subplot(2, 2, 1)
    plt.plot([i for i in range(len(drl_5p))], drl_5p, label="On-Fuzzer", color="red")
    plt.plot([i for i in range(len(ga_5p))], ga_5p, label="GA", color="blue")
    plt.plot([i for i in range(len(rand_5p))], rand_5p, label="Random", color="gray")

    plt.title(scenario +' (5 params)')
    plt.xlabel('Episodes')
    plt.ylabel('Cumulative Violation')
    plt.xticks([i for i in range(30)][::5])
    plt.legend()

    plt.subplot(2, 2, 2)
    plt.plot([i for i in range(len(drl_10p))], drl_10p, label="On-Fuzzer", color="red")
    plt.plot([i for i in range(len(ga_10p))], ga_10p, label="GA", color="blue")
    plt.plot([i for i in range(len(rand_10p))], rand_10p, label="Random", color="gray")

    plt.title(scenario +' (10 params)')
    plt.xlabel('Episodes')
    plt.ylabel('Cumulative Violation')
    plt.xticks([i for i in range(30)][::5])
    plt.legend()

    plt.subplot(2, 2, 3)
    plt.plot([i for i in range(len(drl_15p))], drl_15p, label="On-Fuzzer", color="red")
    plt.plot([i for i in range(len(ga_15p))], ga_15p, label="GA", color="blue")
    plt.plot([i for i in range(len(rand_15p))], rand_15p, label="Random", color="gray")

    plt.title(scenario +' (15 params)')
    plt.xlabel('Episodes')
    plt.ylabel('Cumulative Violation')
    plt.xticks([i for i in range(30)][::5])
    plt.legend()

    plt.subplot(2, 2, 4)
    plt.plot([i for i in range(len(drl_20p))], drl_20p, label="On-Fuzzer", color="red")
    plt.plot([i for i in range(len(ga_20p))], ga_20p, label="GA", color="blue")
    plt.plot([i for i in range(len(rand_20p))], rand_20p, label="Random", color="gray")

    plt.title(scenario +' (20 params)')
    plt.xlabel('Episodes')
    plt.ylabel('Cumulative Violation')
    plt.xticks([i for i in range(30)][::5])
    plt.legend()
    
    # show plot
    plt.subplots_adjust(bottom=.01)
    plt.show()
    plt.clf()
```

```python
import matplotlib.pyplot as plt

def plot_comparison(scenario, ga, drl, rand):
    """
    Plots the comparison of cumulative violations for GA, DRL, and Random algorithms.

    Parameters:
    - scenario: The name of the flight mode/scenario (string).
    - ga: Data for the GA algorithm (list or array).
    - drl: Data for the DRL algorithm (list or array).
    - rand: Data for the Random algorithm (list or array).
    """
    plt.figure(figsize=(10, 6))

    # Plotting the lines
    plt.plot(range(len(ga)), ga, label="GA", color="blue")
    plt.plot(range(len(drl)), drl, label="DRL", color="red")
    plt.plot(range(len(rand)), rand, label="Random", color="gray")

    # Adding title and labels
    plt.title(f'{scenario} - Cumulative Violations Comparison')
    plt.xlabel('Episodes')
    plt.ylabel('Cumulative Violation')
    plt.xticks(range(0, len(ga), 5))
    plt.legend()

    # Display the plot
    plt.show()
    plt.clf()

# Example usage
# Assuming ga_data, drl_data, and rand_data are lists or arrays of data for a specific flight mode
# scenario = 'A.RTL2'
# plot_comparison(scenario, ga_data, drl_data, rand_data)
```
